```python
import requests

URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
page=requests.get(URL)

from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')

# Get main content <div>
poke_content=soup.find(id='mw-content-text')

# Get all <table> elements
poke_tables=poke_content.find_all('table')
```

--- Getting generations 2-8

```python
poke_tables[2].contents[3].contents[5].contents[1].get('href')
```

```python

start_index = 3

generations_dicts = []

for i in range(2,9):
    generation = poke_tables[i]
    gen_dict = {'generationNo': i, 'Pokemons': []}
    for i in range(start_index, len(generation.contents), 2):
        poke_info = generation.contents[i]
        kdex = poke_info.contents[1].text.strip()
        ndex = poke_info.contents[3].text.strip()
        wiki = poke_info.contents[5].contents[1].get('href')
        name = poke_info.contents[7].text.strip()
        type1 = poke_info.contents[9].text.strip()
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            print(f'Pokemon {ndex} {name} is a {type1} & {type2} Pokemon')
            gen_dict['Pokemons'].append({
                'kdex': kdex,
                'ndex': ndex,
                'wiki': wiki,
                'name': name,
                'type1': type1,
                'type2': type2
            })
        else:
            print(f'Pokemon {ndex} {name} is a {type1} Pokemon')
            gen_dict['Pokemons'].append({
                'kdex': kdex,
                'ndex': ndex,
                'wiki': wiki,
                'name': name,
                'type1': type1
            })
        print(wiki)
    generations_dicts.append(gen_dict)

        
```

```python
import json

pokeJsonFile = open('Gen2to8.json', "w")

json.dump(generations_dicts, pokeJsonFile)
```
